```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from collections import Counter
from matplotlib.gridspec import GridSpec
%matplotlib inline

import seaborn as sns
sns.set(color_codes=True)
```

```python
Mac_data = pd.read_csv('Mcdonald .csv')
Mac_data
```

```python
# Q1: Plot graphically which food categories have the highest and lowest varieties.
```

```python

print(Counter(Mac_data['Category']))
```

```python
# grouping our data by the number of values for each ‘Category’ category:
title_category = Mac_data.groupby('Category').agg('count')
title_category
```

```python
#sort the indices and counts for our aggregated types:
type_labels = title_category.Item.sort_values().index 
type_counts = title_category.Item.sort_values()
```

```python
#figure details:
plt.figure(1, figsize=(6,5)) 
the_grid = GridSpec(2, 2)
```

```python
# mapping colour info
cmap = plt.get_cmap('Spectral')
colors = [cmap(i) for i in np.linspace(0, 1, 8)]
```

```python
#plotting pie chart:
plt.subplot(the_grid[0, 1], aspect=1, title='Categories')
type_show_category = plt.pie(type_counts, labels=type_labels, autopct='%1.1f%%', shadow=True, colors=colors, radius=5)
plt.show()
```

## From the above count plot, we find out:

### 1. Max Varieties - Coffee & Tea
### 2. Min Varieties - Salads

## Q4. Which category contributes to the maximum % of Cholesterol in a diet (% daily value)?

```python
per_data = Mac_data.groupby('Category').sum()['Cholesterol (% Daily Value)'].sort_values(ascending = False)
plt.subplots(figsize=(15,6))
plt.plot(Mac_data.Category, Mac_data.Cholesterol)
plt.xlabel('Category')
plt.ylabel('Choestrol')
plt.title('% Of Cholesterol in a diet')
plt.show()
```

## Q5. Which item contributes maximum to the Sodium intake?

```python
per_data = Mac_data.groupby('Item').sum()['Sodium'].sort_values(ascending = False)
print(per_data)
plt.subplots(figsize=(20,10))
plt.plot(Mac_data.Item, Mac_data.Sodium)
plt.xlabel('Item')
plt.ylabel('Sodium')
plt.title('Soudium Intake')
plt.show()
```

## From the above results, Chicken McNuggets(40 piece)' contributes to maximum Sodium intake (3600).
